```python
import os #for directory path and to read file
import shutil #copy,move file
import random #to generate random numbers
from tqdm.notebook import tqdm #to show progress bar
```

```python
#!pwd
```

```python
train_img_path="./yolo_data/images/train/" #training images
train_label_path="./yolo_data/labels/train/" #.txt output of training images
val_img_path="./yolo_data/images/val/" #testing data i/p
val_label_path="./yolo_data/labels/val/" #o/p of testing data
test_path='./yolo_data/test/' #extra external image storage #these all are used for directories
```

```python
random.seed(42) #train_test split
```

```python
def train_testsplit(path,split=0.2):
  files=list(set([name[:-4] for name in os.listdir(path)])) #to get just names from data foleder avoiding .jpg,.txt and using set to avoid duplicate names
  print("Number Of Files",len(files)) #66
  random.shuffle(files)
  test_size=int(len(files)*split) #66*0.2==13.2==>13
  print("Test image size is",test_size)
  train_size=len(files)-test_size #53
  print("Train image size is",train_size)

  #create a directory
  os.makedirs(train_img_path, exist_ok=True)
  os.makedirs(train_label_path, exist_ok=True)
  os.makedirs(val_img_path, exist_ok=True)
  os.makedirs(val_label_path, exist_ok=True) #creating folders if path exists

  #linux command copy command:cp sourcepath destinationpath

  for file in tqdm(files[:train_size]):
    shutil.copy2(path+file+'.jpg',f"{train_img_path}/"+file+'.jpg')
    shutil.copy2(path+file+'.txt',f"{train_label_path}/"+file+'.txt')

  for file in tqdm(files[train_size:]):
     shutil.copy2(path+file+'.jpg',f"{val_img_path}/"+file+'.jpg')
     shutil.copy2(path+file+'.txt',f"{val_label_path}/"+file+'.txt')







```

```python
train_testsplit('/content/drive/MyDrive/Yolov8/data/')
```
