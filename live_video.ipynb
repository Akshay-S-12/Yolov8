```python
import ultralytics
model=ultralytics.YOLO('yolov8n.pt')
```

```python
#live video detection
import cv2
video=cv2.VideoCapture(0) #we can give video file path instead of 0 for webcam
while True:
    suc,img=video.read()
    result=model(img)
    for i in result:
        i.save(filename='result.jpg')
        img=cv2.imread('result.jpg')
        cv2.imshow('Image',img)
    if cv2.waitKey(1) & 0xFF==ord('q'): #press q to stop
        break   
video.release()
cv2.destroyAllWindows()
```
